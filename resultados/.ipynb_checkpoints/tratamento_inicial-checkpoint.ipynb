```python
import pandas as pd
import numpy as np
```

# MMQ junção dos dados

```python
slots_adicionais = ['slot02', 'slot03', 'slot04', 'slot06', 'slot09', 'slot10', 'slot11']
```

```python
dfMMQ = pd.read_excel('comportamentoMMQ_slot01.xlsx')
```

```python
for slot in slots_adicionais:
    df_aux = pd.read_excel('comportamentoMMQ_' + slot +'.xlsx')
    dfMMQ = pd.concat([dfMMQ,df_aux], ignore_index = True)
```

## MP tratamento dos dados

```python
dfMMQ = dfMMQ.drop(columns=dfMMQ.columns[dfMMQ.columns.str.contains('^Unnamed')])
```

```python
dfMMQ['Matriz'] = dfMMQ['Matriz'].str.replace('.mtx', '')
```

```python
dfMMQ
```

```python
dfMMQ2 = dfMMQ[dfMMQ['r2'] != 'erro']
```

```python
dfMMQ2
```

```python
df2MMQ = dfMMQ2.query('r2 > 0.95')
df2MMQ = df2MMQ.query('Metodo == "MP" ')
df2MMQ
```

```python
valores_unicos = df2MMQ['Matriz'].unique()
```
