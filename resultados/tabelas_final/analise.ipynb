```python
def calcular_erro_por_matriz(df, metodo_padrao):
    erros_por_matriz = {}
    matrizes = df['Matriz'].unique()

    for matriz in matrizes:
        matriz_df = df[df['Matriz'] == matriz]
        matriz_df['Tempo_float'] = matriz_df['Tempo'].astype(float)
        matriz_df['Iteracoes_float'] = matriz_df['Iterações'].astype(float)

        erro_iteracoes = np.abs(matriz_df.loc[matriz_df['Aceleração'] == metodo_padrao, 'Iteracoes_float'].values[0] - matriz_df['Iteracoes_float'])/ np.abs(matriz_df.loc[matriz_df['Aceleração'] == metodo_padrao, 'Iteracoes_float'].values[0]) 
        erro_por_matriz = list(erro_iteracoes)
        erros_por_matriz[matriz] = erro_por_matriz

    return erros_por_matriz

metodo_padrao = 'Nenhuma'
errors = calcular_erro_por_matriz(df, metodo_padrao)
print(errors)
```
