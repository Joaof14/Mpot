```python
import pandas as pd
import numpy as np
```

# MMQ junção dos dados

```python
dfMMQ = pd.read_excel('comportamentoMMQ_slot01.xlsx')
```

```python
df_aux = pd.read_excel('comportamentoMMQ_slot11.xlsx')
dfMMQ = pd.concat([dfMMQ,df_aux], ignore_index = True)
```

## MP tratamento dos dados

```python
dfMMQ = dfMMQ.drop(columns=dfMMQ.columns[dfMMQ.columns.str.contains('^Unnamed')])
```

```python
dfMMQ['Matriz'] = dfMMQ['Matriz'].str.replace('.mtx', '')
```

```python
dfMMQ
```

```python
#salvar
dfMMQ.to_excel('tabelas_final/resultado_MMQ_final.xlsx')
```
